**W.1** Implement a function that takes as input three variables, and returns the largest of the three. Do this without using the Python max() function.

The goal of this exercise is to think about some internals that Python normally takes care of for us. All you need is some variables and if statements.

```python
def largest_of_three(a, b, c):
    """Return the largest of three inputs without using max()."""
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# quick tests
print(largest_of_three(3, 7, 5))
print(largest_of_three(-2, -8, -1))
print(largest_of_three(4, 4, 2))
```

```python
# E.2 Pandas cheat sheet / notes

import pandas as pd
import numpy as np

# -----------------------------
# 1. Creating and reading data
# -----------------------------
# Read a CSV file into a DataFrame
# df = pd.read_csv("my_file.csv")

# Example DataFrame for notes:
df = pd.DataFrame({
    "country": ["Jamaica", "Barbados", "Jamaica", "Trinidad"],
    "year": [2020, 2020, 2021, 2021],
    "temp_anomaly": [0.9, 0.8, 1.0, 1.1],
    "rainfall": [1200, 1400, 1100, np.nan]
})

print("Original DataFrame:")
print(df)

# -----------------------------
# 2. Basic inspection
# -----------------------------
print("\nFirst 2 rows with head():")
print(df.head(2))

print("\nColumn names:")
print(df.columns)

print("\nShape (rows, columns):")
print(df.shape)

print("\nData types:")
print(df.dtypes)

# -----------------------------
# 3. Selecting columns
# -----------------------------
# Single column -> Series
print("\nSingle column:")
print(df["country"])

# Multiple columns -> DataFrame
print("\nMultiple columns:")
print(df[["country", "year"]])

# -----------------------------
# 4. Filtering rows
# -----------------------------
# Boolean filtering
print("\nRows where country is Jamaica:")
print(df[df["country"] == "Jamaica"])

print("\nRows where temp anomaly > 0.9:")
print(df[df["temp_anomaly"] > 0.9])

# Multiple conditions: use &, | and wrap each condition in parentheses
print("\nRows where country is Jamaica AND year is 2021:")
print(df[(df["country"] == "Jamaica") & (df["year"] == 2021)])

# -----------------------------
# 5. loc and iloc
# -----------------------------
# .loc -> label-based selection
print("\nUsing .loc to get rows 0:2 and columns country + rainfall:")
print(df.loc[0:2, ["country", "rainfall"]])

# .iloc -> position-based selection
print("\nUsing .iloc to get first 3 rows and first 2 columns:")
print(df.iloc[0:3, 0:2])

# -----------------------------
# 6. Creating new columns
# -----------------------------
# New columns can be made from old ones
df["temp_F_estimate"] = df["temp_anomaly"] * 1.8 + 32

print("\nDataFrame with new column:")
print(df)

# -----------------------------
# 7. Sorting
# -----------------------------
print("\nSort by temp_anomaly ascending:")
print(df.sort_values("temp_anomaly"))

print("\nSort by year descending:")
print(df.sort_values("year", ascending=False))

# -----------------------------
# 8. Summary statistics
# -----------------------------
print("\nMean temperature anomaly:")
print(df["temp_anomaly"].mean())

print("\nSummary statistics:")
print(df.describe(include="all"))

# -----------------------------
# 9. Groupby
# -----------------------------
# Grouping lets you summarize by category
print("\nAverage temp anomaly by country:")
print(df.groupby("country")["temp_anomaly"].mean())

print("\nAverage rainfall by year:")
print(df.groupby("year")["rainfall"].mean())

# -----------------------------
# 10. Pivot tables
# -----------------------------
pivot = df.pivot_table(values="temp_anomaly", index="country", columns="year")
print("\nPivot table of temp anomaly by country and year:")
print(pivot)

# -----------------------------
# 11. Missing data
# -----------------------------
print("\nMissing values in each column:")
print(df.isna().sum())

print("\nDrop rows with missing values:")
print(df.dropna())

print("\nFill missing rainfall with 0:")
print(df["rainfall"].fillna(0))

# -----------------------------
# 12. Important reminders
# -----------------------------
# - Use [] to select columns.
# - Use .loc for names/labels and .iloc for positions.
# - Use boolean masks to filter rows.
# - groupby() is useful for category summaries.
# - sort_values() sorts rows.
# - isna(), dropna(), and fillna() help with missing data.
# - A DataFrame is a table; a Series is a single column.

print("\nCheat sheet complete.")
```
