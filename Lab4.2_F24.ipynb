```python


# 1. For Loop
# Used to iterate over sequences (lists, strings, ranges)

for i in range(5):
    print(i)

# range(start, stop, step)
for i in range(1, 10, 2):
    print(i)

# Loop through a list
numbers = [1, 2, 3, 4]
for num in numbers:
    print(num)

# 2. While Loop
# Runs while condition is True

count = 0
while count < 5:
    print(count)
    count += 1

# 3. Break
# Stops loop early

for i in range(10):
    if i == 5:
        break
    print(i)

# 4. Continue
# Skips current iteration

for i in range(5):
    if i == 2:
        continue
    print(i)

# 5. Enumerate
# Gives index + value

names = ["Aiden", "Corey", "Mom"]
for index, name in enumerate(names):
    print(index, name)

# 6. Nested Loops

for i in range(3):
    for j in range(2):
        print(i, j)

print("Loops are powerful for repeating tasks and analyzing data.")
```
